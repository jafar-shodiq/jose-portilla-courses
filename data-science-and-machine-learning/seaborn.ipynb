# Section 9 Lecture 49: Categorical Plots

```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**Barplot**

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

**Countplot**

```python
sns.countplot(x='sex', data=tips)
```

**Boxplot**

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

**Violinplot**

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

**Stripplot**

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

**Swarmplot**

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

**Factorplot**

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

# Section 9 Lecture 50: Matrix Plots

```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

**Heatmap**

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
```

```python
sns.heatmap(fp, cmap='coolwarm')
```

**Clustermap**

```python
sns.clustermap(fp)
```

# Section 9 Lecture 51: Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**PairGrid**

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

**FacetGrid**

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

```python
g.map(sns.scatterplot, 'total_bill', 'tip')
```

# Section 9 Lecture 52: Regression Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**Lmplot**

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

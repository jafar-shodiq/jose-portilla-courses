# Section 8 Lecture 42: Matplotlib Part 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-')
```

```python
plt.plot(x, y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

**Subplot**

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

**Object-oriented method**

```python
fig = plt.figure()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Big Plot')

axes2.plot(y, x)
axes2.set_title('Small Plot')
```

# Section 8 Lecture 43: Matplotlib Part 2

**Subplot OO**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

**Figure size and DPI**

```python
fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x ,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('./others/my_fig_1.png', dpi=200)
```

**Legend**

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc='upper left')
```

# Section 8 Lecture 44: Matplotlib Part 3

**Plot appearance**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x,
        y,
        color='#FF8C00',
        linewidth=3,
        alpha=0.5,
        linestyle='--',
        marker='o',
        markersize=10,
        markerfacecolor='green',
        markeredgewidth=2,
        markeredgecolor='black')
```

**Plot range**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='red', linewidth=2, linestyle='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```

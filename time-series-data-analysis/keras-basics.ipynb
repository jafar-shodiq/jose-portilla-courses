```python
import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
m = 2
b = 3
x = np.linspace(0, 50, 100)
```

```python
np.random.seed(101)
noise = np.random.normal(loc=0, scale=4, size=len(x))
```

```python
y = 2*x + b + noise
```

```python
plt.plot(x, y, '*')
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4, input_dim=1, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(x, y, epochs=200, verbose=False)
```

```python
loss = model.history.history['loss']
epochs = range(len(loss))

plt.plot(epochs, loss)
```

```python
x_for_predictions = np.linspace(0, 50, 100)

y_preds = model.predict(x_for_predictions)
```

```python
plt.plot(x, y, '*')
plt.plot(x_for_predictions, y_preds, 'r')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y, y_preds)
```
